```python
import pandas_datareader as pdr 
import pandas as pd 
from datetime import datetime 
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
df_tesla.head(4)
```

```python
#plotting of graph
df_tesla['High'].plot(figsize=(12,5))
```

```python
#X limmit and y limmit
df_tesla['High'].plot(xlim=['01-01-2020','09-12-2022'],figsize=(12,5))
```

```python
#y limit
df_tesla['High'].plot(xlim=['2021-01-01','2021-12-09'],ylim=[0,900],figsize=(12,4),ls='--', c='green')
```

```python
index = df_tesla.loc['2020-01-01':'2021-01-01'].index
share_open = df_tesla.loc['2020-01-01':'2021-01-01']['Open']
```

```python
# creating a subplot wrt to open share
import matplotlib.pyplot as plt 
%matplotlib inline

figure, axis =plt.subplots()
axis.plot(index, share_open)
```
